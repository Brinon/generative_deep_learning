```python
import os

import numpy as np
import matplotlib.pyplot as plt

from models import VariationalAutoencoder
from utils.data import load_mnist
from utils.callbacks import step_decay_schedule
```

```python
vae = VariationalAutoencoder.build_default()
```

```python
vae.model.summary()
```

```python
(x_train, y_train), (x_test, y_test) = load_mnist()
```

```python
lr = 0.0005
r_loss_factor = 1000
vae.compile(lr, r_loss_factor)
vae.model.fit(
    x_train, 
    y = x_train,
    batch_size = 32,
    shuffle=True,
    epochs=100,
    callbacks = [
        step_decay_schedule(initial_lr=lr, decay_factor=1, step_size=1)
    ]
)
```
